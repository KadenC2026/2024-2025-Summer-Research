# Collateralized Debt Obligations and the Credit Crisis of 2007
### Background Context
+ **Bubble**: When the price of an asset or a sector rises dramatically, exceeding its fundamental or intrinsic value, and is ultimately unsustainable.
  + Often caused by Speculation, buying without thinking of risk
+ In 2000, Housing prices rise drastically. Due to following reasons:
    + Low Interest rates on Borrowing
    + Increase in **Subprime first mortgage Lending**, which is lending to people with below-average credit scores or limited credit history.
+ As house prices kept on increasing, it became more difficult for first-time buyers to buy houses, so lenders had to relax their lending standards even more.

### Bubble Bursts
As housing prices rise, the demand for the houses decreased, which started decreasing the housing prices.
  + Lower housing prices resulted in borrowers having negative equity (The loan they have to pay back is more than the value of the house they own), which led to foreclosures and an increase in supply of houses, which decreases housing prices.
  + Homes served like an **American Put Option**, meaning the owner of the asset (home) can sell the home anytime at a set strike price (the mortgage they have to pay back).
    + If the house price drops below the mortgage price, the borrower can sell the house to the bank, and forego the mortgage they have to pay. Essentially, the borrower is selling the house for the mortgage price.
      + Ex: Suppose someone bought a house for 300,000 dollars with a loan of 290,000 dollars (The Strike Price).
      + Later, the house might only be worth $200,000 (The market Price).
      + Since market price < Strike Price, the homeowner could simply walk away (default) and give the house to the bank, which removes the mortgage debt.
  + As house prices dropped, borrowers started exercising these "put options," giving the banks undervalued houses.
  + **Securitization**: Securitization is the process of taking mortages, and pooling them together to create new, tradable securities that can be sold to investors.
    + Banks passed on their mortgages to other investors through securitization.
    + Banks no longer cared whether borrowers could repay their mortgages — because the risk was passed on to investors via securitization.
    + This led to a rise in subprime lending. <br>

###Asset-Backed Securities
  + **Asset-Backed Securities (ABS)**: Investment products backed by a pool of underlying assets, such as loans, leases, or receivables. <br> <br>
    1: A portfolio of assets (such as subprime mortgages) is sold by the originators of the assets to a special purpose vehicle (SPV) and the cash flows from the assets are allocated to **tranches**. <br> <br>
    2: There are the senior tranche, the mezzanine tranche, and the equity tranche. As shown below. This is divided as follows: 75 million dollars to the senior tranche, 20
million dollars to the mezzanine tranche, and 5 million dollars to the equity tranche. The senior
tranche is promised a return of 6%, the mezzanine tranche is promised a return of
10%, and the equity tranche is promised a return of 30%. <br> <br>

  3: Over time, borrowers repay their loans (with interest). Those cash flows go to:

    + First pay the senior tranche investors (6% return),
    + Then mezzanine (10%),
    + And finally equity (30%) — **only if there’s money left.**

Example 1: Let’s say the $100 million asset pool performs poorly:

 + Borrowers default, and only $85 million + Interest is recovered.

 + The $15 million loss hits the equity tranche first, then goes to the Mezzanine:

| Tranche           | Principal | Loss |
| :---------------- | :------: | ----: |
| Equity        |   5 Million   | Full \$5 million Loss |
| Mezzanine           |   20 million   | $10 Million Loss with no return |
| Senior    |  75 million   | Fully repaid + 6% return |

Senior Tranche gets paid back first, and is the safest Tranche while the Equity tranche has the possibility of getting big returns (30%), but also has the biggest risk.

Example 2: Let’s say the $100 million asset pool performs well:

 + Borrowers pay everything back, 100 million + interest


| Tranche           | Principal | Loss |
| :---------------- | :------: | ----: |
| Equity        |   5 Million   | Fully repaid + 30% Interest |
| Mezzanine           |   20 million   | Fully repaid + 10% Interest |
| Senior    |  75 million   | Fully repaid + 6% Interest |

If the asset performs well, the equity tranche gets the biggest return.

## ABS CDOs
Finding investors for the Mezzanine Tranche was difficult, so financial engineers created an ABS from the mezzanine tranches of ABSs that were created from subprime mortgages.
+ **ABS CDO**: An ABS of the Mezzanine tranche, with a new Equity, Mezzanine, and Senior Tranche

### Problem related to CDO: <br>
Suppose a CDO is constructed from a pool of 100 loans, each with notional 1 million dollars. The equity tranche covers losses from $0%$–$5%$, the mezzanine from $5%$–$15%$, and the senior from $15%$–$100%$.

(a) If 8 loans default by time $T$, each with zero recovery, compute the loss for each tranche.

(b) Generalize: write a formula for the loss to the $[a,b]$ tranche as a function of total portfolio loss $L(T)$.

#### Solution:
a) If 8 loans default, than the total portfolio loss is 8 million. The equity tranche will cover the first 5 million of losses, and the next 3 million will be covered by the mezzanine tranche. The Senior Tranche has no losses. <br>
b) We need to take into account parameters:
 - We start with L(t)-a. If this is negative, we need to set the losses to zero. If this is greater than b-a, we set the loss to b-a. <br>
 Putting this into mathematical notation:
 #### $\mathrm{Tranche\ Loss} = \min\!\{\max(L(t)-a,\,0),\,b - a\}.$
